#About the dataset

Bigbasket is the largest online grocery supermarket in India. Was launched somewhere around in 2011 since then they've been expanding their business. Though some new competitors have been able to set their foot in the nation such as Blinkit etc. but BigBasket has still not loose anything - thanks to ever expanding popular base and their shift to online buying.

###Features of the dataset

This dataset contains 10 attributes with simple meaning and which are described as follows:

index - Simply the Index!

product - Title of the product (as they're listed)

category - Category into which product has been classified

sub_category - Subcategory into which product has been kept

brand - Brand of the product

sale_price - Price at which product is being sold on the site

market_price - Market price of the product

type - Type into which product falls

rating - Rating the product has got from its consumers

description - Description of the dataset (in detail)

While doing FE, "discount" can be created as (marketprice - saleprice)/ market_price * 100 which will help in getting what consumers are getting better here!

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('/content/drive/MyDrive/BigBasket/BigBasket Products.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.drop('index', axis=1, inplace=True)
```

<b>Rating has many null values.To deal with this we can take mean rating of each category then fill Null ratings in accordance with the catagory they belong to.<b>

```python
 df['category'].unique()
```

```python
df['rating'].fillna(df.groupby("category")['rating'].transform("mean"),inplace=True)
```

```python
# Null Value Percentage Per column
percent_missing = df.isnull().sum() * 100 / len(df)
missing_value_df = pd.DataFrame({'column_name': df.columns,
                                 'percent_missing': percent_missing})
missing_value_df
```

Percentage Missing is reduced drastically

<b>The Average Sale Price, Rating, Market_Price Per category<b>

```python
df.groupby('category').mean().sort_values('rating', ascending = False)
```

<b> Ranking the categories according to the rating <b>

```python
avg_rating = df.groupby('category').mean().reset_index().sort_values('rating', ascending = False).head(9)
fig = px.bar(avg_rating, x='category', y='rating', template = 'plotly_dark', title='Ranking of Catagories WRT Rating', color='rating', height=800)
fig.update_xaxes(tickangle=90, rangeselector_font_size=8)
fig.show()
```

<b> Ranking of Categories with respect to average market price <b>

```python
avg_df = df.groupby('category').mean().reset_index()
```

```python
avg_market_price = avg_df.sort_values('market_price', ascending = True)
fig = px.bar(avg_market_price, x='market_price', y='category', template = 'plotly_dark', title='Ranking of Catagories WRT AVG market price', color='market_price', height=800)
fig.update_xaxes(tickangle=90, rangeselector_font_size=8)
fig.show()
```

<b> Ranking of Categories with respect to average sales price<b>



```python
avg_sales_price = avg_df.sort_values('sale_price', ascending = True)

fig = px.bar(avg_sales_price, x='sale_price', y='category', template = 'plotly_dark', title='Ranking of Catagories WRT average sales price', color='sale_price', height=800)

fig.show()
```

<b>Scatter plot of market price V/s Sale Price<b>

```python
fig = px.scatter(avg_sales_price,x = 'sale_price',y = 'market_price',color='sale_price',size='sale_price',hover_data=['category', 'sale_price', 'market_price', 'rating'],
                 template = 'plotly_dark', 
                 title='Scatter Plot of cross Market Price and Sale Price')
fig.show()
```
